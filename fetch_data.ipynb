```python
# DOW30 Companies
DOW30_csv = "DOW30.csv"  
df_DOW30 = pd.read_csv(DOW30_csv)
DOW30_tickers = df_DOW30['Symbol'].tolist()
```

```python
DOW30_tickers
```

```python
# download data for DOW30 - this method produces multiindexed data, which is not ideal
# 2023 Year
#DOW30_2023 = yf.download(tickers=DOW30_tickers, start="2024-01-05", end="2024-01-06")
```

```python
# Retrieve data for DOW30 Companies for 2023 
DOW30_df_list = list()
for ticker in DOW30_tickers:
    data = yf.download(ticker, group_by="Ticker", start="2023-01-01", end="2023-12-31")
    data['ticker'] = ticker
    DOW30_df_list.append(data)

DOW30_df = pd.concat(DOW30_df_list)

DOW30_df.to_csv('DOW30_data.csv')
```

```python
# check for missing values
DOW30_df.isna().sum()
```

```python
# Retrieve data for 2023 
NASDAQ_df_list = list()         
for ticker in NASDAQ_tickers:
    data = yf.download(ticker, group_by="Ticker", start="2023-01-01", end="2023-12-31")
    data['ticker'] = ticker
    NASDAQ_df_list.append(data)

NASDAQ_df = pd.concat(NASDAQ_df_list)

NASDAQ_df.to_csv('NASDAQ_data.csv')
```
